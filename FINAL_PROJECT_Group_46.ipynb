```r
library(tidyverse)

options(repr.matrix.max.rows= 8)

School_hdi_data <- read_tsv("/home/jupyter/Group_46_Final_Project/Expected_years of schooling_(years).tsv",
                           col_names = FALSE,
                            skip =6)
School_hdi_data

#This was doen to skip the first 6 columns.

School_hdi_data_1 <- School_hdi_data %>%
    slice(1:(n()-9))
#This was done to get rid of all the bottom rows that had irrelevant/ extra information

School_hdi_data_2 <- School_hdi_data_1 %>%
        select(X1:X62)%>%
        filter(X1 != "NA") 
#This was done to remove any colums that had NA values for the HDI as they are not going to contribute to answering the question.

School_hdi_data_3 <- School_hdi_data_2[, colSums(is.na(School_hdi_data_2)) < nrow(School_hdi_data_2)]
#This was done to get rid of all the columns that had NA as values for all the rows.

drop <- c("X62")

School_hdi_data_4 = School_hdi_data_3[,!(names(School_hdi_data_3) %in% drop)]
#This was doen to delete the 62nd row

School_hdi_data_5 <- School_hdi_data_4[-c(1),]                         
#This was done to make the first row the column names

cols <- c("HDI_Rank","Country","1990","1991", "1992", "1993", "1994","1995", "1996", "1997", "1998","1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007",  "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015","2016", "2017", "2018", "2019")
colnames(School_hdi_data_5 ) <- cols
School_hdi_data_5
#This was done to give the colums column names

School_hdi_data_6 <- pivot_longer(School_hdi_data_5, cols = "1990":"2019",
                 names_to = "Years",
                values_to = "Years_of_Schooling")
School_hdi_data_6 
#This step tidies the data into 4 colums that all have one variable.

glimpse(School_hdi_data_6)
#The glimpse provides a small look into what the table of data is made up of

School_hdi_data_7 <- School_hdi_data_6 %>%
        select("HDI_Rank", "Country", "Years", "Years_of_Schooling")%>%
        filter(HDI_Rank != "..", Years_of_Schooling != ".." )
#This step allows the exclusion of rows with ".." values

```

```r
School_hdi_data_8 <- School_hdi_data_7 %>%
        mutate(HDI_Rank = as.numeric(HDI_Rank))%>%
        mutate(Years_of_Schooling = as.numeric(Years_of_Schooling))
School_hdi_data_8
#This changes the HDI_Rank into a numerical value which allows for further analysis



School_hdi_data_9 <- School_hdi_data_8%>%
                select(-Country,-Years)
School_hdi_data_9
#This omits the columns that do not contribute to our analysis

library(dplyr)

School_hdi_data_10 <- School_hdi_data_9 %>%
                    mutate(ranges = cut(HDI_Rank,breaks = c(1, 66, 119, 156, 189),labels = c("very_high","high","medium", "low"), include.lowest = TRUE)) %>%
                    group_by(ranges) %>%
                    summarise(Number_of_rows= n())%>%
                    rename( HDI_ranges = ranges)
School_hdi_data_10
#This creates a table that places rows into a certain range, and counts the numbers of rows in that range. 
```

```r
set.seed(624) #ONLY WRITE THIS ONCE

library(repr)
library(tidymodels)

School_hdi_split <- initial_split(School_hdi_data_9, prop = 0.75, strata = HDI_Rank)
School_hdi_split

School_hdi_train <- training(School_hdi_split)
School_hdi_train

School_hdi_test <- testing(School_hdi_split)
#This splits the data set into training and testing data
```
