```r
library(tidyverse)
```

```r


School_hdi_data <- read_tsv("/home/jupyter/Expected years of schooling (years) (2).tsv",
                           col_names = FALSE,
                            skip =6)
School_hdi_data

```

```r
School_hdi_data_1 <- School_hdi_data %>%
    slice(1:(n()-9))
               
School_hdi_data_1
        
```

```r
School_hdi_data_2 <- School_hdi_data_1 %>%
        filter(X1 != "NA")%>%
        select()
School_hdi_data_2
                
```

```r
School_hdi_data_3 <- School_hdi_data_2[, colSums(is.na(School_hdi_data_2)) < nrow(School_hdi_data_2)]
School_hdi_data_3
```

```r
drop <- c("X62")

School_hdi_data_4 = School_hdi_data_3[,!(names(School_hdi_data_3) %in% drop)]
School_hdi_data_4
```

```r
names(School_hdi_data_4) <- as.matrix(School_hdi_data_4[1,])
School_hdi_data_4 <- School_hdi_data_4[-1,]
School_hdi_data_4[] <- lapply(School_hdi_data_4, function (x) type.convert(as.character(x)))
School_hdi_data_4
```
